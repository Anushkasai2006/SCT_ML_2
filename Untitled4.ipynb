```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gradio as gr
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
url = "https://gist.githubusercontent.com/pravalliyaram/5c05f43d2351249927b8a3f3cc3e5ecf/raw/Mall_Customers.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Encode Gender: Male = 0, Female = 1
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

# Select features for clustering
X = df[['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Scale the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method - Optimal Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Age',
    y='Spending Score (1-100)',
    hue='Cluster',
    palette='Set2',
    data=df,
    s=100
)
plt.title("Customer Segments (Age vs Spending Score)")
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```

```python
import gradio as gr

def predict_cluster(gender, age, income_in_lakhs, score):
    # Convert Gender
    gender_encoded = 0 if gender == "Male" else 1

    # Convert INR Lakhs to k$ (1 lakh INR  1.2k$)
    income_kd = income_in_lakhs * 1.2

    # Prepare input and scale
    input_data = [[gender_encoded, age, income_kd, score]]
    input_scaled = scaler.transform(input_data)
    cluster = kmeans.predict(input_scaled)[0]

    # Segment descriptions
    segment_names = {
        0: " Young High Spenders - Energetic, shop frequently, love offers",
        1: " Wealthy but Cautious - High earners, spend carefully",
        2: " Balanced Shoppers - Average income and spending",
        3: " Mature Mid Spenders - Older, regular, moderate buyers",
        4: " Impulsive Spenders - Low income but love shopping"
    }

    return f"Segment: {segment_names[cluster]}"



# Build Gradio UI
ui = gr.Interface(
    fn=predict_cluster,
    inputs=[
        gr.Radio(["Male", "Female"], label="Gender"),
        gr.Number(label="Age"),
        gr.Number(label="Annual Income (in  Lakhs)"),
        gr.Number(label="Spending Score (1-100)")
    ],
    outputs="text",
    title=" Mall Customer Cluster Predictor",
    description="Enter customer details to find their segment (based on Gender, Age, Income, Spending Score)"
)

ui.launch()
```
